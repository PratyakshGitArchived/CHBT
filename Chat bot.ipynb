```
from nltk.chat.util import Chat, reflections
```

```
pairs = [
         ['my name is (.*)',['hi %1']],
         ['(hi|hello|hey)',['hey there','hi there','whats up']],
         ['(.*) in (.*) is fun',['%1 in %2 is indeed fun']],
         ['(.*)(location|city) ?',['london, UK']],
         ['(.*) created you ?',['Pratyaksh did using NLTK']],
         ['how is the weather in (.*) ?',['the weather in %1 is Pleasent']],
         ['(.*)help(,*)',['I can help you']],
         ['(.*) your name ?',['my name is Temp 1']]
]
```

```
reflections
```

```
my_dummy_reflections = {
    'go' : 'gone',
    'hello' : 'hey there'
}
```

```
chat = Chat(pairs, my_dummy_reflections)
chat.converse()
```
